# Q3 Goal: Determine Probability and Distribution

```python
import re
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt 
```

### a. Find the probability of the word "data" occuring at each line

```python
new_dict = {}
ind=0

# I copied the sentences in to a text file called abc.txt:
filename = 'abc.txt'    
with open(filename, "r") as fp:
    contents = fp.readlines()
    #join the list of text in a single string
    joint_words = ''.join(contents)
    # removes punctuation in the string using re.sub()
    re_joint_words = re.sub(r'[^\w\s]','', joint_words)
    # make all words into lower case
    joint_words_lower=re_joint_words.lower()
    #find the number of count for the word "data analytics"
    data_analytics_count = joint_words_lower.count('data analytics')
    #split into list of lines for probability calculation
    fp2 = joint_words_lower.splitlines()
    
    for line in fp2:
        words = line.split()
        total_words_inline = len(words)
        data_count_inline = words.count('data')
        
        prob_inline = data_count_inline/total_words_inline
        ind = ind+1
        print('''The probability of the word "data" occuring at line %s is %f''' % (ind, prob_inline) )
        #build word library
        for word in words:

            if word not in new_dict:
                new_dict[word] = 1
            else:
                new_dict[word] += 1
```

### b. Plot distribution of each distinct word

#### Frequency distribution of each distinct word

```python
#sort the dictionary based on value
sorted_dict=dict(sorted(new_dict.items(), key=lambda item: item[1]))

## Conversion to Pandas series easier plotting
all_fdist = pd.Series(sorted_dict)

## plot the bar distribution
fig, ax = plt.subplots(figsize=(80,30))
all_plot = sns.barplot(x=all_fdist.index, y=all_fdist.values, ax=ax)
plt.title("Frequency Distribution of Distinct Words", fontsize=70)
plt.xticks(rotation=50, fontsize=40)
plt.yticks(fontsize=60);
plt.xlabel('Word', fontsize=50)
plt.ylabel('Count', fontsize=50)
plt.show()
```
